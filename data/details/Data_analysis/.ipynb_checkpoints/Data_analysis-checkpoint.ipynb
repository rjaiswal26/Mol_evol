# Visualization

```python
for i in list(df.columns)[1:]:
    sn.boxplot(
        data=pd.melt(pd.concat([df.loc[:, df.columns == i],df1.loc[:,  df1.columns == i]],axis=1,keys= ['Pre_Cal', 'Cal'])),
        x='variable_1',
        y='value',
        hue= 'variable_0'
    )
    plt.savefig(str(i)+'_.jpg')
    plt.clf()
```
